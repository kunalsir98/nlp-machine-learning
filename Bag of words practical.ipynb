```python
import pandas as pd 
messages=pd.read_csv('sms.csv',
                     sep='\t',names=['labels','message'])
```

```python
messages
```

```python
# data cleaning and preprocessing 
import re 
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)

```

```python
corpus
```

create bag of words model 


```python
from sklearn.feature_extraction.text import CountVectorizer

#for binary bag of words enable binary =true
cv=CountVectorizer(max_features=2500,binary=True)
```

```python
X=cv.fit_transform(corpus).toarray()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

#for binary bag of words enable binary =true
cv=CountVectorizer(max_features=500,binary=True,ngram_range=(2,3))
X=cv.fit_transform(corpus).toarray()
```

```python
cv.vocabulary_
```

```python
y=pd.get_dummies(messages['labels'])
```

```python
y
```

```python
y=y.iloc[:,0].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
X_train
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam=MultinomialNB()
```

```python
spam.fit(X_train,y_train)
```

```python
y_pred=spam.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
accuracy_score=(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
